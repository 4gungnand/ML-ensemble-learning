```python
# Definisikan fungsi untuk mengisi nilai null
def isi_nilai_null(df, kolom_kategorikal):
    for kolom in kolom_kategorikal:
        df[kolom] = df[kolom].fillna('Unknown')
    for kolom in df.columns:
        if kolom not in kolom_kategorikal:
            df[kolom] = df[kolom].fillna(0)

# Definisikan list kolom kategorikal
kolom_kategorikal = ['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature', 'SaleType', 'SaleCondition']

# Isi nilai null dengan 'Unknown' atau 0 untuk nilai null besar
isi_nilai_null(df, kolom_kategorikal)
isi_nilai_null(df2, kolom_kategorikal)

# Hapus baris dengan nilai null kecil (kurang dari 10)
df = df.dropna()
df2 = df2.dropna()
```

```python
# Untuk hyperparameter tuning dan rmse score
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression, Lasso, ElasticNet, Ridge
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor, ExtraTreesRegressor, AdaBoostRegressor, GradientBoostingRegressor, VotingRegressor, StackingRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Split data menjadi X dan y
X = df.drop('SalePrice', axis=1)
y = df['SalePrice']

# Split data menjadi train dan test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1)
```

Linear Regression

```python
# 1. Linear Regression
linear_reg = LinearRegression()
linear_param_grid = {'copy_X':[True, False]}
linear_reg_grid = GridSearchCV(linear_reg, linear_param_grid, cv=5, scoring='neg_root_mean_squared_error')
linear_reg_grid.fit(X_train, y_train)
linear_reg_best = linear_reg_grid.best_params_
linear_req_test = linear_reg_grid.predict(X_test)
linear_req_rmse = np.sqrt(mean_squared_error(y_test, linear_req_test))
```

Lasso Regression

```python
# 2. Lasso Regression
lasso_reg = Lasso()
lasso_param_grid = {'alpha': [0.001, 0.01, 0.1, 1]}
lasso_reg_grid = GridSearchCV(lasso_reg, lasso_param_grid, scoring='neg_mean_squared_error', cv=5)
lasso_reg_grid.fit(X_train, y_train)
lasso_reg_best = lasso_reg_grid.best_estimator_
lasso_reg_pred = lasso_reg_best.predict(X_test)
lasso_reg_rmse = np.sqrt(mean_squared_error(y_test, lasso_reg_pred))
```

ElasticNet Regression

```python
# 3. ElasticNet Regression
elasticnet_reg = ElasticNet()
elasticnet_param_grid = {'alpha': [0.001, 0.01, 0.1, 1], 'l1_ratio': [0.25, 0.5, 0.75]}
elasticnet_reg_grid = GridSearchCV(elasticnet_reg, elasticnet_param_grid, scoring='neg_mean_squared_error', cv=5)
elasticnet_reg_grid.fit(X_train, y_train)
elasticnet_reg_best = elasticnet_reg_grid.best_estimator_
elasticnet_reg_pred = elasticnet_reg_best.predict(X_test)
elasticnet_reg_rmse = np.sqrt(mean_squared_error(y_test, elasticnet_reg_pred))
```

Ridge Regression

```python
# 4. Ridge Regression
ridge_reg = Ridge()
ridge_param_grid = {'alpha': [0.001, 0.01, 0.1, 1]}
ridge_reg_grid = GridSearchCV(ridge_reg, ridge_param_grid, scoring='neg_mean_squared_error', cv=5)
ridge_reg_grid.fit(X_train, y_train)
ridge_reg_best = ridge_reg_grid.best_estimator_
ridge_reg_pred = ridge_reg_best.predict(X_test)
ridge_reg_rmse = np.sqrt(mean_squared_error(y_test, ridge_reg_pred))
```

```python
# Tampilkan hyperparameter terbaik dari semua model
print('Linear Regression Best Parameter: ', linear_reg_best)
print('Lasso Regression Best Parameter: ', lasso_reg_best)
print('ElasticNet Regression Best Parameter: ', elasticnet_reg_best)
print('Ridge Regression Best Parameter: ', ridge_reg_best)

# Tampilkan rmse score dari semua model
print('Linear Regression RMSE: ', linear_req_rmse)
print('Lasso Regression RMSE: ', lasso_reg_rmse)
print('ElasticNet Regression RMSE: ', elasticnet_reg_rmse)
print('Ridge Regression RMSE: ', ridge_reg_rmse)
```

- Regresi ElasticNet memiliki RMSE terendah (35071.43) dengan parameter terbaik alpha=0.1 dan l1_ratio=0.25.

Bagging Regressor

```python
# 5. Bagging Regressor
bagging_reg = BaggingRegressor()
bagging_reg.fit(X_train, y_train)
bagging_reg_pred = bagging_reg.predict(X_test)
bagging_reg_rmse = np.sqrt(mean_squared_error(y_test, bagging_reg_pred))
```

Random Forest Regressor

```python
# 6. Random Forest Regressor
random_forest_reg = RandomForestRegressor()
random_forest_reg.fit(X_train, y_train)
random_forest_reg_pred = random_forest_reg.predict(X_test)
random_forest_reg_rmse = np.sqrt(mean_squared_error(y_test, random_forest_reg_pred))
```

Extra Tree Regressor

```python
# 7. Extra Trees Regressor
extra_trees_reg = ExtraTreesRegressor()
extra_trees_reg.fit(X_train, y_train)
extra_trees_reg_pred = extra_trees_reg.predict(X_test)
extra_trees_reg_rmse = np.sqrt(mean_squared_error(y_test, extra_trees_reg_pred))
```

AdaBoost Regressor

```python
# 8. AdaBoost Regressor
adaboost_reg = AdaBoostRegressor()
adaboost_reg.fit(X_train, y_train)
adaboost_reg_pred = adaboost_reg.predict(X_test)
adaboost_reg_rmse = np.sqrt(mean_squared_error(y_test, adaboost_reg_pred))
```

Gradient Boosting Regressor

```python
# 9. Gradient Boosting Regressor
gradient_boosting_reg = GradientBoostingRegressor()
gradient_boosting_reg.fit(X_train, y_train)
gradient_boosting_reg_pred = gradient_boosting_reg.predict(X_test)
gradient_boosting_reg_rmse = np.sqrt(mean_squared_error(y_test, gradient_boosting_reg_pred))
```

```python
# 10. Voting Regressor
voting_reg = VotingRegressor(estimators=[('bg', bagging_reg), ('rf', random_forest_reg), ('gbr', gradient_boosting_reg)]
                                , weights=[2,5,3])
voting_reg.fit(X_train, y_train)
voting_reg_pred = voting_reg.predict(X_test)
voting_reg_rmse = np.sqrt(mean_squared_error(y_test, voting_reg_pred))
```

```python
# 11. Stacking Regressor
stacking_reg = StackingRegressor(estimators=[('bg', bagging_reg), ('rf', random_forest_reg), ('gbr', gradient_boosting_reg)]
                                , final_estimator=LinearRegression())
stacking_reg.fit(X_train, y_train)
stacking_reg_pred = stacking_reg.predict(X_test)
stacking_reg_rmse = np.sqrt(mean_squared_error(y_test, stacking_reg_pred))
```

```python
# Tampilkan rmse score dari semua model
print('Bagging Regressor RMSE: ', bagging_reg_rmse)
print('Random Forest Regressor RMSE: ', random_forest_reg_rmse)
print('Extra Trees Regressor RMSE: ', extra_trees_reg_rmse)
print('AdaBoost Regressor RMSE: ', adaboost_reg_rmse)
print('Gradient Boosting Regressor RMSE: ', gradient_boosting_reg_rmse)
print('Voting Regressor RMSE: ', voting_reg_rmse)
print('Stacking Regressor RMSE: ', stacking_reg_rmse)
```

Gradient Boosting Regressor juga memiliki RMSE yang rendah.
